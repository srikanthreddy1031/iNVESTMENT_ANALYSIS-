```python
#TASK 5
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
# Replace 'your_dataset.csv' with the path to your dataset
df = pd.read_csv('data_set2.csv')

# List of columns containing reasons for each investment
investment_columns = ['Reason_Equity', 'Reason_Mutual', 'Reason_Bonds', 'Reason_FD']

# Dictionary to store aggregated reasons and themes
investment_summary = {}
overall_reasons = []

# Step 2 & 3: Explore Reasons Columns, Aggregate Reasons, and Summarize Reasons
for column in investment_columns:
    # Step 2: Explore Reasons Column
    reasons_column = df[column].dropna().str.lower()
    
    # Add reasons to the overall list
    overall_reasons.extend(reasons_column)
    
    # Split the reasons into individual words
    words = reasons_column.str.split(expand=True).stack()
    
    # Count the frequency of each word
    word_counts = words.value_counts()
    
    # If you have predefined themes, you can categorize reasons
    themes = {
        'growth': ['growth', 'increase', 'profit','High ','Interest' ,'Rates','capital' ,'appreciation'],
        'safety': ['safe', 'security', 'low risk', 'Safe ','investment','Risk Free','liquidity'],
        'diversification': ['diversify', 'variety', 'spread',' diversification'],
        'returns': ['return', 'income', 'earn','Fixed','Returns','Better','Returns','Assured', 'Returns','Dividend','Benefits',' Incentives'],
        'knowledge': ['know', 'understand', 'familiar'],
    }
    
    # Initialize a dictionary to hold theme counts
    theme_counts = {theme: 0 for theme in themes}
    
    # Count occurrences of each theme
    for theme, keywords in themes.items():
        for keyword in keywords:
            theme_counts[theme] += reasons_column.str.contains(keyword).sum()
    
    # Store summary for the current investment
    investment_summary[column] = {
        'word_counts': word_counts,
        'theme_counts': theme_counts
    }

# Step 4: Overall Summary
overall_reasons_series = pd.Series(overall_reasons)

# Split the reasons into individual words
overall_words = overall_reasons_series.str.split(expand=True).stack()

# Count the frequency of each word
overall_word_counts = overall_words.value_counts()

# Count occurrences of each theme
overall_theme_counts = {theme: 0 for theme in themes}
for theme, keywords in themes.items():
    for keyword in keywords:
        overall_theme_counts[theme] += overall_reasons_series.str.contains(keyword).sum()

# Function to plot word counts
def plot_word_counts(word_counts, title):
    word_counts.head(10).plot(kind='bar', color='skyblue')
    plt.title(title)
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45, ha='right')
    plt.show()

# Function to plot theme counts
def plot_theme_counts(theme_counts, title):
    pd.Series(theme_counts).plot(kind='bar', color='lightgreen')
    plt.title(title)
    plt.xlabel('Themes')
    plt.ylabel('Occurrences')
    plt.xticks(rotation=45, ha='right')
    plt.show()

# Display the detailed summary for each investment and plot visualizations
for investment, summary in investment_summary.items():
    print(f"\nSummary of reasons for {investment}:")
    print("\nMost common words:")
    print(summary['word_counts'].head(10))
    plot_word_counts(summary['word_counts'], f'Most Common Words in {investment}')
    
    print("\nTheme counts:")
    for theme, count in summary['theme_counts'].items():
        print(f"{theme.capitalize()}: {count} occurrences")
    plot_theme_counts(summary['theme_counts'], f'Theme Counts in {investment}')

# Display the overall summary and plot visualizations
print("\nOverall Summary of reasons for all investments:")
print("\nMost common words:")
print(overall_word_counts.head(20))
plot_word_counts(overall_word_counts, 'Most Common Words in All Investments')

print("\nOverall theme counts:")
for theme, count in overall_theme_counts.items():
    print(f"{theme.capitalize()}: {count} occurrences")
plot_theme_counts(overall_theme_counts, 'Theme Counts in All Investments')
```

```python
# TASK  8
# Import necessary libraries
import pandas as pd

# Step 1: Load the dataset
# Replace 'your_dataset.csv' with the path to your dataset
df = pd.read_csv('data_set2.csv')

# Step 2: Examine the Investment Duration Column
# Replace 'Investment_Duration' with the actual column name in your dataset
column_name = 'Duration'

# Display the first few rows of the dataset to ensure the column exists
print("First few rows of the dataset:")
print(df.head())

# Step 3: Convert duration ranges to numerical values
# Define a function to convert duration ranges to average numerical values
def convert_duration(duration):
    if pd.isna(duration):
        return None
    duration = duration.strip().lower()
    if duration == 'less than 1 year':
        return 0.5
    elif duration == '1-3 years':
        return 2
    elif duration == '3-5 years':
        return 4
    elif duration == 'more than 5 years':
        return 6
    else:
        return None

# Apply the function to the duration column
df[column_name + '_numeric'] = df[column_name].apply(convert_duration)

# Step 4: Calculate the average investment duration
average_duration = df[column_name + '_numeric'].mean()

# Display the average duration
print(f"\nThe average investment duration is: {average_duration:.2f} years")

# Optional: Display the transformed dataset to verify changes
print("\nTransformed dataset with numeric durations:")
print(df.head(45))
```

```python
df[column_name] = df[column_name].apply(convert_duration)

# Count the frequency of each duration
duration_counts = df[column_name].value_counts()

# Plotting a pie chart
plt.figure(figsize=(6, 6))
plt.pie(duration_counts, labels=duration_counts.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Investment Duration Distribution')
plt.show()
```

```python
#TASK 10
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset (replace 'your_dataset.csv' with your actual dataset path)
df = pd.read_csv('data_set2.csv')

# Display the first few rows of your dataset to understand its structure
print("First few rows of the dataset:")
print(df.head())

# Select relevant columns for correlation analysis
columns_to_analyze = ['age', 'Duration_Numeric', 'Expect_Numeric']

# Calculate correlation matrix
correlation_matrix = df[columns_to_analyze].corr()

# Print correlation matrix
print("\nCorrelation Matrix:")
print(correlation_matrix)

# Visualize correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()

# Visualize correlations using pairplot (optional)
sns.pairplot(df[columns_to_analyze])
plt.suptitle('Pairplot of Age, Investment Duration, and Expected Returns', y=1.02)
plt.show()
```
